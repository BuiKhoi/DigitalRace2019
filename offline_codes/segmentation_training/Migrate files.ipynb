```python
import os
import numpy as np
```

```python
root = './Data/Nhatrang/Training_data/val/'
base = './Data/Nhatrang/Training_data/train/'
count = 0
for dirr in os.listdir(root):
    if '.npz' in dirr:
        rand = np.random.random()
        if rand > 0.5:
            os.rename(root + dirr, base + dirr)
            count += 1
            print('Moved {} files'.format(count))
```
