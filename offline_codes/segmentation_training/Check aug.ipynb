```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
rand = np.random.randint(3630)
count = 0
for dirr in os.listdir('./Data/SPKT_LHU/Training/train/'):
    if 'npz' in dirr:
        if count < rand:
            count += 1
            continue
        else:
            print(dirr)
            print(dirr)
            temp = np.load('./Data/SPKT_LHU/Training/train/' + dirr)
            img = temp['img']
            anno = temp['anno']
            break
plt.figure(figsize=(10, 10))
plt.subplot(2, 2, 1)
plt.imshow(img.astype('float'))
plt.title('Input image')
# img = np.expand_dims(img, 0)
# pred = run_model.predict(img)

plt.subplot(2, 2, 2)
plt.imshow(anno[:, :, 0].astype('float'), 'gray')
plt.title('Label road')

plt.subplot(2, 2, 3)
plt.imshow(anno[:, :, 1].astype('float'), 'gray')
plt.title('Label Line')

plt.subplot(2, 2, 4)
plt.imshow(anno[:, :, 2].astype('float'), 'gray')
plt.title('Label sign')
```
